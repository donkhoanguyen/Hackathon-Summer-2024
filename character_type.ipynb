```python
import pandas as pd
```

```python
# Specify the file path
file_path = 'data/Tab_delimited_text/Hackathon2024.ATAC.txt'
atac = pd.read_csv(file_path, sep='\t')
atac.head()
```

```python
# Create a mapping for chromosomes
mapping = {f'chr{i}': f'{i}' for i in range(1, 23)}
mapping.update({'chrX': 'X', 'chrY': 'Y'})
atac[['chromosome', 'peak_start_pos', 'peak_end_pos']] = atac['peak'].str.split('[-]', expand=True)

# Replace values in the DataFrame
atac['chromosome'] = atac['chromosome'].map(mapping)

atac.head(10)
```

```python
# Define the chromosome to type mapping
def map_chromosome_type(chromosome):
    metacentric = ['1', '3', '16', '19', '20']
    submetacentric = ['2', '4', '5', '6', '7', '8', '9', '10', '11', '12', '17', '18', 'X']
    acrocentric = ['13', '14', '15', '21', '22', 'Y']
    
    if chromosome in metacentric:
        return 'Metacentric'
    elif chromosome in submetacentric:
        return 'Submetacentric'
    elif chromosome in acrocentric:
        return 'Acrocentric'
    else:
        return 'Unknown'

# Apply the mapping to the atac['chromosome'] column
atac['chromosome_type'] = atac['chromosome'].apply(map_chromosome_type)
atac.head()
```

```python
# Specify the file path
file_path = 'data/Tab_delimited_text/Hackathon2024.RNA.txt'
rna = pd.read_csv(file_path, sep='\t')
rna.head()
```

```python
rna['gene']
```

```python
# Define the prefix to gene type mapping
def map_gene_type(gene):
    protein_coding_prefixes = []  # No specific prefixes for protein-coding genes
    lncRNA_prefixes = ['LINC', 'MIR', 'AL', 'AC']
    small_rna_prefixes = ['SNORD', 'SNORA', 'RNU', 'MIR']
    pseudogene_prefixes = ['AC', 'AL', 'RP', 'AP']

    if any(gene.startswith(prefix) for prefix in lncRNA_prefixes):
        return 'Long non-coding RNA'
    elif any(gene.startswith(prefix) for prefix in small_rna_prefixes):
        return 'Small non-coding RNA'
    elif any(gene.startswith(prefix) for prefix in pseudogene_prefixes):
        return 'Pseudogene'
    else:
        return 'Protein-coding'

# Apply the mapping to the atac['gene'] column
atac['gene_type'] = atac['gene'].apply(map_gene_type)
```
